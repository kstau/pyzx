```python
import sys; sys.path.append('..')
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import pyzx as zx
from pyzx.flow import *
from pyzx.flow_extract import *
from pyzx.flow_rules import *
import random
```

```python
# c = zx.Circuit.load('../circuits/feyn_bench/qasm/tof_3.qasm').to_basic_gates()
random.seed(1344)
g_orig = zx.generate.cliffordT(4,30,p_t=0.40) # Generate the circuit as a ZX-diagram
c = zx.Circuit.from_graph(g_orig) # Convert it to a sequence-of-gates representation
g = c.to_graph(backend='mbqc')
zx.simplify.id_simp(g)
zx.simplify.to_gh(g)
zx.simplify.spider_simp(g)
zx.simplify.id_simp(g)
g.normalize()
print(c.stats())
zx.draw(g, labels=True)
```

```python
for v in g.vertices():
    if not v in g.inputs() and not v in g.outputs():
        if g.phase(v) == Fraction(1,1) or g.phase(v) == 0:
            g.set_mtype(v, MeasurementType.X)
        elif g.phase(v) == Fraction(1,2) or g.phase(v) == Fraction(3,2):
            g.set_mtype(v, MeasurementType.Y)
        else:
            g.set_mtype(v, MeasurementType.XY)
```

```python
zx.draw(g, labels=True)
```

```python
# pivot(g, 22,32)
# pivot(g, 17,18)
for i in range(0,4):
    if i % 2 == 0:
        lcomp(g, 22)
    else:
        lcomp(g, 32)
    zx.draw(g, labels=True)
    print(zx.compare_tensors(g_orig,g))
```
