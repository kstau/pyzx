```python
for seed in [1249,529,586,484]:
    random.seed(seed)
    g = zx.generate.cliffordT(4,50,p_cnot=0.5,p_t=0.4) # Generate the circuit as a ZX-diagram
    g = zx.Graph(backend='mbqc').from_json(g.to_json())
    zx.simplify.spider_simp(g)
    zx.simplify.to_gh(g)
    zx.simplify.id_simp(g)
    zx.simplify.spider_simp(g)
    convert_to_extractable_graph(g)
    assign_pauli_xy_measurements(g)
    g.normalize()
    print("original graph")
    zx.draw(g, labels=True)
    for i in range(0,4):
        vertex = random.choice([v for v in g.mvertices() if not v in g.minputs() and not v in g.moutputs()])
        lcomp(g, vertex)
        print("lcomp on",vertex)
    print("lcomp graph")
    gcopy = g.copy()
    zx.draw(gcopy, labels=True)
    extract_from_pauli_flow(gcopy)
```
